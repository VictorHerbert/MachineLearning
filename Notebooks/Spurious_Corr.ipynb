```python
#v = pd.DataFrame(data={'year' : [str(s) for s in range(1990,2020)]} )

v=pd.DataFrame(columns=['series'] + [s for s in range(1990,2020)])

for j in range(1,5):
    print(j)
    
    s = req.get('https://tylervigen.com/page?page='+str(j)).text

    df = pd.read_html(s)

    for i in range(2,len(df)-1,2):    
        d = df[i].loc[:2,:]
        d = d.loc[:,:d.shape[1]-2]
        d.loc[0,0] = '0'

        cols = d.loc[0].astype(int)
        cols[0] = 'series'
        d.columns = cols

        d = d.loc[1:]

        v = v.append(d)
    
v.to_csv('scrap.csv')
```
