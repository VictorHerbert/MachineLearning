```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
from functools import reduce
import numpy as np
import seaborn as sn
from  matplotlib import pyplot as plt
```

```python
p = {}
#4492

stk = []

for i in range(5):
    print(i*10,'%')

    df = pd.read_html(
        req.get('https://tylervigen.com/page?page='+str(i)).text
    )

    for j in range(2,len(df)-2,2):
        d = df[j].loc[:2,:]

        ss = list(d[0].unique()[1:])

        

        years = list(d.loc[0][1:-1].astype('int32'))
        for k in [0,1]:
            s = ss[k]
            v = list(d.loc[k+1][1:-1].astype('float'))

            if not s in p:
                p[s] = {}
            for l in range(len(years)):
                p[s][years[l]] = v[l]

v = pd.DataFrame.from_dict(p).dropna(axis=1,how = 'all')
v.to_csv('scrap.csv')
```
